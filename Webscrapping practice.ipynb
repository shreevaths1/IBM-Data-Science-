### Wikipedia dataset

```python
url = 'https://en.wikipedia.org/wiki/World_population'
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
data = requests.get(url).text
# data
```

```python
r = requests.get(url)
print(r.headers['content-type'])
```

```python
soup = BeautifulSoup(data,'html.parser')
soup
```

```python
tables = soup.find_all('table')
len(tables)
```

```python
for l in tables:
    if l.caption != None:
        print(l.caption.text)
```

```python
for index,table in enumerate(tables):
    if 'World historical and predicted populations' in str(table):
        table_index = index
        break
print(table_index)
```

```python
print(tables[table_index].prettify())
```

```python
tables[table_index].tbody.tr
```

```python
for row in tables[table_index].tbody.find_all('tr'):
    col = row.find_all('td')
    if col!=[]:
        print(row)
        
```
